# Benchmarking // Whole Life Carbon

```python
import pandas as pd

# List of variables to compute statistics for
variables = [
    'GHG_A123_m2a_harm', 
    'GHG_A45_m2a_harm',
    'GHG_B1234_m2a_harm',
    'GHG_B67_m2a_harm',
    'GHG_C12_m2a_harm',
    'GHG_C34_m2a_harm',]

# Filter the dataset
filtered_df = df[(df['site_region_geoscheme'] == 'Western Europe') 
                 & (df['bldg_use_subtype'] == 'Single family house')
                 #& (df['bldg_energy_class_general'] == 'New Standard') 
                 #& (df['bldg_struct_type'] == 'massive brick')
                 ]

# Initialize an empty DataFrame to store the results
results_df = pd.DataFrame()

# Loop through each variable to compute the descriptive statistics
for var in variables:
    descriptive_stats = filtered_df[var].describe(percentiles=[.25, .5, .75])
    percentiles = descriptive_stats[['count', '25%', '50%', '75%']]
    percentiles_df = percentiles.to_frame().T
    results_df = pd.concat([results_df, percentiles_df])

results_df
```
